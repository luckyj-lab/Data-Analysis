## Data Exploration by Visualization

### Introduction
In this project, we shall be working on a data set about the outcomes of student who graduated from college between 2010 to 2012.The original data was released by American Community Survey but was later cleaned by FiveThirthEight and released on the Github repo.
Using visualization, we can answer questions such as :
- Do students in more popular majors make more money?
- How many majors are predominantly male and which are predominantly female?
- Which category of majors has the most students?

```python
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from  pandas.plotting import scatter_matrix as sm
%matplotlib inline
```

```python
#Read csv rfile
recent_grads = pd.read_csv("recent-grads.csv")
```

```python
#First row formated as a table
recent_grads.iloc[0]
```

```python
#View head of dataframe
recent_grads.head(5)
```

```python
#View Tail of dataframe
recent_grads.tail(5)
```

```python
#View summary statistics
recent_grads.describe()
```

```python
#View number of rows and columns in data set
raw_data_count = len(recent_grads.index)
print(raw_data_count)
```

```python
#Drop missing data 
recent_grads = recent_grads.dropna()
```

```python
#View new shape of data frame
cleaned_data_count = len(recent_grads.index)
print(cleaned_data_count)
```

From the above results, only one row contained missing data.Number of rows reduced from 173 to 172 

## Pandas, Scatter Plot

```python
#Exploring the relationship between Sample_size and Median
fig,ax = plt.subplots()
ax.scatter(recent_grads['Sample_size'], recent_grads['Median'])
ax.set_xlabel("Sample_size")
ax.set_ylabel("Median")
plt.show()
```

```python
#Exploring the relationship between Sample_size and Unemployment_rate
fig,ax = plt.subplots()
ax.scatter(recent_grads['Sample_size'], recent_grads['Unemployment_rate'])
ax.set_xlabel("Sample_size")
ax.set_ylabel("Unemployment_rate")
plt.show()
```

```python
#Exploring the relationship between Full_time and Median
fig,ax = plt.subplots()
ax.scatter(recent_grads['Full_time'], recent_grads['Median'])
ax.set_xlabel("Full_time")
ax.set_ylabel("Median")
plt.show()
```

```python
#Exploring the relationship between ShareWomen and Unemployment_rate
fig,ax = plt.subplots()
ax.scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'])
ax.set_xlabel("ShareWomen")
ax.set_ylabel("Unemployment_rate")
plt.show()
```

```python
#Exploring the relationship between Men and Median
fig,ax = plt.subplots()
ax.scatter(recent_grads['Men'], recent_grads['Median'])
ax.set_xlabel("Men")
ax.set_ylabel("Median")
plt.show()
```

```python
#Exploring the relationship between women and median
fig,ax = plt.subplots()
ax.scatter(recent_grads['Women'], recent_grads['Median'])
ax.set_xlabel("Women")
ax.set_ylabel("Median")
plt.show()
```

## Pandas, Bar Plots

```python
#Lets compare the percentage of women from the first ten rows and last ten rows of recent_grads
recent_grads[:10].plot.bar(x="Major",y="ShareWomen")
recent_grads[-10:].plot.bar(x="Major",y="ShareWomen")
```

From the above plot,its clear that overall fewer women are graduating from the engineering degrees than the social fields of study.Amongst the first ten field is the Astronomyy which contains over 50 percent of recent graduates

```python
#Lets compare unemployment rate of first ten and last ten rows 
recent_grads[:10].plot.bar(x="Major",y="Unemployment_rate")
recent_grads[-10:].plot.bar(x="Major",y="Unemployment_rate")
```

From the graphs above, its clear that the unemployment rate is lowest for astromy which probably explains why most women go into that field of study.
Overall, the unemployemnt rate is low for the fields of engineering but for nuclear engineering since its not very used.
